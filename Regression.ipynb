# Regression Project

```python
from sklearn.datasets import make_regression
```

```python
features = 20
X,Y = make_regression(n_samples=3000,n_features=features,n_informative=12,noise=50)
```

```python
cols = [f'x{i+1}' for i in range(features)]
cols
```

```python
import pandas as pd
X = pd.DataFrame(X,columns=cols)
X.head()
```

```python
Y = pd.DataFrame(Y,columns=['target'])
Y.head()
```

```python
X.info()
```

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
num_pipe = Pipeline(steps=[('imputer',SimpleImputer(strategy='mean')),
                           ('scaler',StandardScaler())])
```

```python
X_pre = num_pipe.fit_transform(X)
X_pre  = pd.DataFrame(X_pre,columns=X.columns)
X_pre.head()
```

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(X_pre,Y,test_size=0.2,random_state=21)
```

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV
import numpy as np
params = {'alpha':np.arange(0.1,30,0.1)}
params
```

```python
model = Ridge()
gscv = GridSearchCV(model,param_grid=params,cv=5,scoring='neg_mean_squared_error')
gscv.fit(xtrain,ytrain)
```

```python
best_model = gscv.best_estimator_
best_model
```

```python
ypred_tr = best_model.predict(xtrain)
ypred_ts = best_model.predict(xtest)
```
